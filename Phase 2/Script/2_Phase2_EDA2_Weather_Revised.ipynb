"codesum" feature is having multi option categorical feature, we will need to do one hot encoding.To do that - 

1. We first finding out possible number of Unique values - 29 in this case
2. Then will create extra columns for one hot encoding

```python
lstCodesum = list(df_weather.codesum.unique())
#lstCodesum
codesum_set = set()
for codes in lstCodesum:
  lst = str(codes).split(" ") #list (str((codes.split(" "))))
  #print(codes)
  for ele in lst:
    codesum_set.add(ele)

codesum_list = list(codesum_set)
len(codesum_set)
#Total number of Unique values in codesum column
```

Complist: This holds exsisting and newly added columns to have one hot encoding

```python
y = pd.get_dummies(codesum_list, prefix='codesum')
y.columns

Complist = list(df_weather.columns) + list(y.columns)
```

Creating a fresh dataframe with revised list of columns that is Complist & doing one hot encoding

Removing base codesum column (string)

```python
df_weather2 = df_weather.reindex(columns = Complist, copy=True)
#df_weather2 = df_weather2.iloc[:5]
rw = 0
for index, row in df_weather2.iterrows(): 
    chklist = str(row['codesum']).split(" ")
    for cde in codesum_list:
      if cde in chklist:
        col = "codesum_"+cde
        df_weather2.iat[rw, df_weather2.columns.get_loc(col)] = 1
    rw += 1
      

# Removing column name 'codesum'
df_weather2.drop(['codesum'], axis = 1, inplace=True)
df_weather2.head(5) 
```

Creating station wise Median Dataframe, which will be used to fillna

```python
df_weather_median = df_weather3.groupby(['station_nbr']).median()
df_weather_median
```

There are few station where we have NaN in Median dataframe.
Now using the first Median variable that is df_weather_median and filling NaN with Median of df_weather_median

```python
df_weather_median2 = df_weather_median.copy()
df_weather_median2.loc[df_weather_median2['depart'].isnull(),'depart'] = df_weather_median2.loc[df_weather_median2['depart'].isnull(),'depart'].fillna(df_weather_median.depart.median())
df_weather_median2.loc[df_weather_median2['wetbulb'].isnull(),'wetbulb'] = df_weather_median2.loc[df_weather_median2['wetbulb'].isnull(),'wetbulb'].fillna(df_weather_median.wetbulb.median())
df_weather_median2.loc[df_weather_median2['snowfall'].isnull(),'snowfall'] = df_weather_median2.loc[df_weather_median2['snowfall'].isnull(),'snowfall'].fillna(df_weather_median.snowfall.median())

df_weather_median2.loc[df_weather_median2['stnpressure'].isnull(),'stnpressure'] = df_weather_median2.loc[df_weather_median2['stnpressure'].isnull(),'stnpressure'].fillna(df_weather_median.stnpressure.median())
df_weather_median2.loc[df_weather_median2['sealevel'].isnull(),'sealevel'] = df_weather_median2.loc[df_weather_median2['sealevel'].isnull(),'sealevel'].fillna(df_weather_median.sealevel.median())
df_weather_median2
```

Filling up missing values with Station specific Median values using df_weather_median2

```python
df_weather4 = df_weather3.copy()

columnList = ["tmax", "tmin", "tavg", "depart", "dewpoint", "wetbulb", "heat", "cool", "snowfall", "preciptotal", "stnpressure", "sealevel", "resultspeed", "resultdir", "avgspeed"]
for station_nbr in range(1,21):
  for clmn in columnList:
    condition = (df_weather4[clmn].isnull()) & (df_weather4['station_nbr'] == station_nbr)
    df_weather4.loc[condition,clmn] = df_weather4.loc[condition,clmn].fillna(df_weather_median2.iloc[station_nbr-1][clmn])


```

Filling up One Hot encoding null values with 0. So that the columns only contains boolean values

```python
df_weather4.update(df_weather3[['codesum_BCFG',	'codesum_BLDU',	'codesum_BLSN',	'codesum_BR',	'codesum_DU',	'codesum_DZ',	'codesum_FG',	'codesum_FG+',	'codesum_FU',	'codesum_FZDZ',	'codesum_FZFG',	'codesum_FZRA',	'codesum_GR',	'codesum_GS',	'codesum_HZ',	'codesum_MIFG',	'codesum_PL',	'codesum_PRFG',	'codesum_RA',	'codesum_SG',	'codesum_SN',	'codesum_SQ',	'codesum_TS',	'codesum_TSRA',	'codesum_TSSN',	'codesum_UP',	'codesum_VCFG',	'codesum_VCTS',	'codesum_nan']].fillna(0))
df_weather4
```

Droping Sunset & Sunrise columns

```python

df_weather5 = df_weather4.copy()
df_weather5.drop(['sunrise','sunset'], axis = 1,inplace=True)
```

Now the weather dataframe is cleaned

Creating Different features - Day, Month, Quarter, Holiday

```python
df_weather6 = df_weather5.copy()
df_weather6["date"] = pd.to_datetime(df_weather6["date"])
df_weather6["Day"] = df_weather6.date.dt.day
df_weather6["Month"] = df_weather6.date.dt.month
df_weather6["Quarter"] = df_weather6.date.dt.quarter

from pandas.tseries.holiday import USFederalHolidayCalendar as calendar

dr = pd.date_range(start='2012-01-01', end='2014-10-31')
cal = calendar()
holidays = cal.holidays(start=dr.min(), end=dr.max())

df_weather6['Holiday'] = df_weather6['date'].isin(holidays)
```
